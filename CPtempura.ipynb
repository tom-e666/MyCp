3. Theme
Thiết lập theme giúp cá nhân hóa trải nghiệm, có thể giúp nhìn code tốt hơn, trông chuyên nghiệp hơn
Bạn có thể chọn giữa theme màu hoặc dùng hình ảnh làm theme.
Mình đặt ảnh anime làm nền VScode, không tệ chút nào.

Các bạn có thể tìm theme cần thiết trong mục extension (ảnh bên dưới). Các theme ở dạng add-on cài dặt thêm. 
Để đặt ảnh nền, bạn cần add-on hỗ trợ việc này. Mình dùng add-on có tên là Background. Bạn cần tải về trước khi tiếp tục thiết lập. Hình ảnh là add-on dầu tiên trong danh sách nhé.

Để đặt ảnh nền:
 1. Ấn tổ hợp Ctrl + Shift + P (hoặc chuột phải và chọn control pallete)
 2. gõ "background", các thếp lập của add-on background sẽ được đề xuất.
 3. Chọn Background: Configuration
 Sau đó, bạn có thể thiết lập cho từng panel của vscode. Mình hài lòng chỉ với thiết lập cho Window
 4. Chọn File và chọn file/folder ảnh cần thiết. Có thể cần restart VSCode để áp dụng.
 5. Opacity: Bước này là cần thiết vì ảnh quá rõ làm code khó nhìn. Chọn Background: Configuration : Opacity và đặt thành 0.9  Bạn có thể điều chỉnh phù hợp với mong muốn.
 
 
 

4. Github Copilot: (free)
Github là công cụ được dùng để quản lí mã nguồn phần mềm. Phát triển phần mềm rất phức tạp yêu cầu một chương trình quản lí mã nguồn (một phần mềm có hàng ngàn cập nhật trong nhiều giai đoạn khác nhau và chỉ một số được sử dụng chính thức). Sử dụng Git là kĩ năng cần thiết sau này khi bạn cần phát triển phần mềm hoặc có nhu cầu chia sẻ code cho người khác.
Github cho ra mắt Github Copilot, một chatbot chạy trên GPT4-turbo chuyên viết code. Copilot này tương tự chatgpt 3.5. Nó có thể nhìn thấy code bạn đang viết, generate code và chỉnh sửa. Nó chỉ trả lời câu hỏi liên quan code và những nội dung khác sẽ không có phản hồi. Copilot chỉ dùng tiếng anh nên không biết tiếng anh sẽ là bất lợi. 

Copilot được thiết kế để hạn chế các bạn lạm dụng giải các bài tập, đừng mong chờ nó giải đúng các bài tập liên quan đến CP nhé.

Github Copilot tính phí 10$/month, trong thời gian là sinh viên sẽ được cung cấp miễn phí. Bạn cần tài khoản microsoft sinh viên, bằng chứng vật lí (thẻ sinh viên) và làm theo hướng dẫn từ trang sau: https://education.github.com/pack
Mình không hướng dẫn thêm, các bạn tìm hiểu qua youtube nhé.
PS: mình submit 30 lần mới thành công, chủ yếu là form thông tin cung cấp cho github không chính xác 100% với thông tin trên Microsoft Student Account.

5. Giao diện 3 khung
 Màn hình code thông dụng gồm 3 bảng (panel): file .cpp,
 file inputf.in và outputf.out

Mình sẽ hướng dẫn đưa cả 3 lên màn hình vscode.
    1. tạo các file cần thiết: file cpp bất kì, inputf.in, outputf.out
    2. Góc trên bên trái, mở View->Editor Layout-> Split Left.
    3. Tiếp tục chuột trái vào bảng bên phải , chọn View-> Editor Layout-> Split Down.
 
 Bạn đã có giao diện 3 ô cần thiết.
 4. Chuột trái vào mỗi panel, ấn Ctrl + O và mở các file cần thiết

 Bố cục truyền thống có inputf.in góc trên bên phải, kề dưới là outputf.out
 Bên trái là file Cpp.
5. Lưu bố cục
Chọn File-> Save As Workspace và đặt tên cho workspace.
Khi mở nhiều cửa sổ VSCode, workspace sẽ không tự duy trì, để lấy lại bố cục, bạn cần mở file workspace được lưu trong ổ đĩa. Cách nhanh nhất là từ File, lướt đến ô recent.







6. I/O qua file 
Để chuyển hướng nhập xuất (I/O) của một chương trình Cpp, bạn có 2 lựa chọn: Một là sử dụng lệnh thực thi freopen (hoặc thay đổi lớp stream: iofstream), Hai là chỉnh sửa lệnh thực thi, muốn biết sâu hơn thì tự tìm hiểu nhé. 

Trước hết, các thao tác này áp dụng lên chỉ một folder, mình khuyến khích các bạn dùng một folder duy nhất chứa tất cả code .cpp, chủ yếu là tiết kiệm thời gian thiết lập này nọ.

1. Xác định file tasks.json có trong folder chưa, file này sẽ được tự động tạo khi chạy chương trình cpp bất kì, nghĩa là nếu không có, hãy viết và chạy 1 file cpp đơn giản nhé.
2. Dán nội dung bên dưới vào tasks.json rồi Save
Vậy là xong.
Đôi khi có lỗi, tự debug nhé.

```python
{
    "version": "2.0.0",
    "tasks": [
        {
           "type": "cppbuild",
            "label": "C/C++: g++.exe build active file",
            "command": "",
            "args": [
                "copy",
                "\"${file}\"",
                "${workspaceFolder}\\jspwTest.cpp",
                "&&",
                "g++",
                "jspwTest.cpp",
                "-o",
                "jspwTest",
                "&&",
                "jspwTest",
                "<",
                "inputf.in",
                ">",
                "outputf.out",
                "&&",
                "del",
                "jspwTest.exe",
                "&&",
                "del",
                "jspwTest.cpp"
            ],
            "presentation": {
                "echo": false,
                "reveal": "silent",
                "focus": false
            },
            "group": {
                "kind": "build",
                "isDefault": true
            },
            "problemMatcher": {
                "owner": "cpp",
                "fileLocation": [
                    "relative",
                    "${workspaceRoot}"
                ],
                "pattern": {
                    "regexp": "^(.*):(\\d+):(\\d+):\\s+(warning|error):\\s+(.*)$",
                    "file": 1,
                    "line": 2,
                    "column": 3,
                    "severity": 4,
                    "message": 5
                }
            }
        },
        
           
    ]
}
```

7.Thay phím tắt

Ctrl + F5 dùng để thực hiện thao tác Run Without Debug. Vì không có Debug nên thời gian thực thi sẽ nhanh hơn.
F5 đôi khi không thuận tay vì xa vị trí đặt tay, mình quen dùng tổ hợp Ctrl + B trong Sublime Text, tổ hợp này dễ làm hardcore hơn. Bạn hãy thử nghiệm theo nhu cầu của bản thân nhé.

1. Mở Command Pallete (Ctrl + Shift + P hoặc chuột phải)
2. Tìm Preferences: Open Keyboard Shortcuts
3. Tìm Debug: Start without debugging trong hộp thoại mới
4. khi tìm thấy ô kể trên, bên trái có kí tự bút chì, trỏ chuột và phím bạn mong muốn (mình là Ctrl B)


8. Mội vài tips khác
 Terminal: 
    chạy trên vscode thì terminal sẽ xuất hiện liên tục, đôi khi phân tán sự tập trung. Mình chưa biết các tắt terminal. Tạm thời bạn có thể thu nhỏ kích thước terminal (nếu ẩn terminal thì nó lại xuất hiện mỗi lần chạy code)
 Format code:
    Code CP không chú trọng thẩm mỹ, nhưng để tự động hóa, bạn có thể sử dụng add-on có tên prettier, mỗi lần thực thi code thì code sẽ được format cho gọn gàng
 Debug console:
    dân chuyên có thể dùng debug console (cerr, clog) để in ra thông tin cần thiết mà không gián đoạn đến output truyền thống (cout). Thay đổi của mình có thể làm debug console không hoạt động, các bạn tự nghiên cứu nhé.
Ẩn file .exe
    mỗi file cpp được thực thi sẽ đi kèm một file .exe, bạn có thể ẩn nó (không cần thiết). Trong khi mở folder, chọn File-> Preferences-> Settings. Trong file Settings tìm file:exclude và ấn button add partern. GÕ **/*.exe
